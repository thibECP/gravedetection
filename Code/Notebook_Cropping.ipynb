# Cropping all images
Crops all images in folder ../images/ and puts them in subfolder ../images/Cropped/

```python
#Makes sure that the subfolder Cropped/ exists in images/, if not create it
directory = '../images/Cropped/'
if not os.path.exists(directory):
    os.makedirs(directory)
```

```python
for imageName in os.listdir('../images/'):
    if 'jpg' in imageName:
        imagePath = '../images/'+imageName
        image = cv2.imread(imagePath)
        imageCropped = cropCemetery(image)
        
        cv2.imwrite('../images/Cropped/'+imageName,imageCropped)
```
